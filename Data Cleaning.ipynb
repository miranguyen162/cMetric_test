```python
#Tạo list tên nghệ sỹ để bổ sung các giá trị thiếu trong cột Target và cột Artist
name = list(mon['Artist'].value_counts().index)
name.remove('All')
#Loại bỏ bản ghi Quốc Trung để chỉ xem xét mức độ thảo luận đối với các nghệ sỹ có biểu diễn
name.remove('Quốc Trung')

#Thêm các giá trị định dạng tên nghệ sỹ để lọc tên một cách chính xác
name.append('LuTreChau')
name.append('Lũ Trẻ Châu')
name.append('LTC')
name.append('Tiên Tiên')
name.append('Tien Tien')
name.append('tiên tiên')
name.append('tien tien')
name.append('Lộn Xộn')
name.append('Hoang Thuy Linh')
name.append('hoang thuy linh')
name.append('hoàng thùy linh')
name.append('Xinh Xô')
name.append('xinh xô')
name.append('Xinh Xo')
name.append('xinh xo')
name.append('Vinh Khuat')
name.append('vinh khuất')
name.append('vinh khuat')
name.append('Hyuk Oh')
name.append('HyukOh')
name.append('Hyuk-Oh')
name
```

```python
#Lọc list Target NULL
tarnull = mon[mon['Target'].isnull()]
tarnull
```

```python
#Bổ sung tên Nghệ sỹ vào các giá trị mà Content có đề cập tới tên nghệ sỹ
for i, v in enumerate(tarnull['Content']):
    for j in range (len(v)):
        for k in name:
            if v[j:(j+len(k))] == str(k):
                tarnull['Target'].iloc[i] = 'Nghệ sỹ'
```

```python
#Update new Target vào dataset ban đầu, update đc 206 rows
mon.update(tarnull['Target'])
mon.info()
```

```python
#cho các giá trị còn lại là 'Ngoại mục'
ngoaimuc = mon[mon['Target'].isnull()]
ngoaimuc['Target'].fillna('Ngoại mục',inplace=True)
```

Làm sạch giá trị NULL cột Artist

```python
#Lọc list Artist NULL
artnull = mon[mon['Artist'].isnull()]
artnull
```

```python
mon['Artist'].value_counts()
```

```python
#Fill các giá trị mà Content có đề cập tên nghệ sỹ
for i, v in enumerate(artnull['Content']):
    num = 0
    for j in range (len(v)):
        for k in name:
            if v[j:(j+len(k))] == str(k):
                artnull['Artist'].iloc[i] = k
                num += 1
    if num > 1:
        artnull['Artist'].iloc[i] = 'All'
```
